```python
from helpers import load_obj
from cascaded import AnalyticCascadedResult
from graph_finding import get_best_perf
import matplotlib.pyplot as plt
from graphs import draw_graph
from networkx import Graph
from numpy import linspace
from os import getcwd
from alcpack import local_complementation
from decoder_class import CascadeDecoder
```

```python
best_graphs = get_best_perf(4, 9, 0.99)
transmissions = linspace(0, 1)
for key, value in best_graphs.items():
    g = Graph()
    g.add_edges_from(value[1][0])
    # draw_graph(g)
    print(value)
    edges, spcr, xr, yr, zr, xyr = value[1]
    r = AnalyticCascadedResult([[spcr, xr, yr, zr, xyr]], [0, 0, 0, 0, 0, 0])
    for d in range(1, 5):
        plt.plot(transmissions, [r.get_spc_prob(t, d) for t in transmissions])
    plt.plot(transmissions, transmissions, '--')
    plt.legend(list(range(1, 5)))
    plt.show()


```
