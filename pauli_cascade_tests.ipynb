```python
from graphs import gen_ring_graph, gen_fullyconnected_graph, gen_rgs_graph, draw_graph, gen_crazy_graph
from decoder_class import CascadeDecoder
import numpy as np
import matplotlib.pyplot as plt
```

```python
bases = ['x', 'y', 'z']
g = gen_crazy_graph(3, 3)
draw_graph(g)

g_dec = CascadeDecoder(g)
for b in bases:
    g_dec.build_tree(basis=b, printing=False)
transmissions = np.linspace(0.001, 0.999)
logical_transmissions = [[g_dec.success_prob_outcome_list(t, depolarizing_noise=0, ec=False, basis=base) for t in transmissions] for base in bases]
plt.plot(transmissions, logical_transmissions[0])
plt.plot(transmissions, logical_transmissions[1])
plt.plot(transmissions, [(logical_transmissions[0][i] + logical_transmissions[1][i])/2 for i in range(len(logical_transmissions[0]))])

plt.plot((0, 1), (0, 1), 'k--')
plt.xlabel('Physical transmission probability')
plt.ylabel('Logical transmission probability')
plt.show()
```
